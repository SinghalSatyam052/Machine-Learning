Support Vector Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create a synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y 
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 10 )
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
#Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## OTHER KERNELS

```python
lin = SVC(kernel = 'linear')
lin.fit(X_train,y_train)
#Prediction
y_pred1 = lin.predict(X_test)
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
rbf = SVC(kernel = 'rbf')  #by default rbf
rbf.fit(X_train,y_train)
#Prediction
y_pred3 = rbf.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
poly = SVC(kernel = 'poly')
poly.fit(X_train,y_train)
#Prediction
y_pred2 = poly.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigm = SVC(kernel = 'sigmoid')
sigm.fit(X_train,y_train)
#Prediction
y_pred4 = svc.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```

## Hyperparameter tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

#defining parameter range
params = {'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001]
           }
```

```python
grid = GridSearchCV(SVC(),param_grid = params,refit =True,cv=5,verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python

#Prediction
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test,y_pred_grid))
print(confusion_matrix(y_test,y_pred_grid))
```
